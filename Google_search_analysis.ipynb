```python
from pytrends.request import TrendReq
```

```python
from matplotlib import pyplot as plt
```

```python
trends=TrendReq()
```

```python
print(trends)
```

```python
print(trends.build_payload(kw_list=["web development"]))
```

```python
data=trends.interest_by_region()
print(data)
```

```python
data=data.sort_values("web development" , ascending=False)
#print(data)
data=data.head(10)
print(data)
```

```python
data.reset_index().plot(x="geoName" ,y="web development",figsize=(15,5),kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
print(data)
```

```python
print(data.build_payload(kw_list=['web development']))
```

```python
data = data.interest_over_time()
print(data)
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
data['web development'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for web development', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
